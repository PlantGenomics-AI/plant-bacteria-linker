Algoritmo DBSCAN com dados sint√©ticos

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados sint√©ticos
X = np.random.rand(100, 2)

# Aplicando o DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=5)
labels = dbscan.fit_predict(X)

# Visualizando os resultados
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```

Algoritmo Hierarquico com dados sint√©ticos

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados sint√©ticos
X = np.random.rand(100, 2)

# Aplicando o Clustering Hier√°rquico
agglo = AgglomerativeClustering(n_clusters=3)
labels = agglo.fit_predict(X)

# Visualizando os resultados
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Hierarchical Clustering')
plt.show()
```

Algoritmo K-Means com dados sint√©ticos

```python
import sklearn
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Gerando dados sint√©ticos
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Aplicando o K-means
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Visualizando os clusters
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')  # Centroides
plt.title('K-means Clustering')
plt.show()
```
